```python
import matplotlib.pyplot as plt
import torch
from torchvision import transforms as T
from dataset.FastPatchExtractor import FastPatchDataset
```

```python
dims = (40, 40)
path = '../data/simple/more_diverse_simple.png'
```

```python
dataset = FastPatchDataset(path, dims, padding=False, transforms=T.Compose([T.Lambda(lambda im: im.reshape(1, *dims)/255.0)]))
```

```python
dataset[10].shape
```

```python
random = torch.randint(0, len(dataset), (100000,))
sum_max = torch.mul(*dims)
patch_sample = [int(torch.sum(dataset[i])) for i in random]
print('sample length', len(patch_sample))
plt.hist(patch_sample, bins=range(1, torch.mul(*dims)-1), edgecolor='black', alpha=0.7)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Frequency Occurrences')
plt.show()
```
