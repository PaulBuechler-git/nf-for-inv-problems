```python
from PIL import Image
from matplotlib import pyplot as plt
from torchvision import transforms
from operators.gaussian_blur import GaussianBlur
```

```python
transformPIL = transforms.Compose([transforms.PILToTensor()])
```

```python
file = "../data/set12/train/01.png"
image = transformPIL(Image.open(file))
```

```python
pytorch_transforms = transforms.Compose([transforms.Lambda(lambda img: img / 255.0)])
normalized_img = pytorch_transforms(image)
normalized_img.shape
```

```python
k = 7
std = 3
```

```python
blur = GaussianBlur(k, std)
blurred_image = blur(normalized_img)
blurred_image.shape
```

```python
# Create a figure with two subplots arranged horizontally
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

# Display the first image on the left subplot
axes[0].imshow(normalized_img.squeeze(), cmap='gray')
axes[0].set_title('Clean')
axes[0].axis('off')

# Display the second image on the right subplot
axes[1].imshow(blurred_image.squeeze(), cmap='gray')
axes[1].set_title(f'Gaussian Blur with k={k} and std={std}')
axes[1].axis('off')

# Adjust spacing between subplots
plt.tight_layout()

# Show the figure
plt.show()
```
