```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import make_interp_spline
from scipy.signal import argrelextrema

# Generate an ill-conditioned distribution using a negative exponential function
np.random.seed(0)
ill_conditioned_data = np.random.exponential(scale=7.0, size=10000)

# Take the logarithm to stabilize the distribution
log_data = np.log(ill_conditioned_data)

# Create a histogram for the ill-conditioned data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
hist, bins, _ = plt.hist(ill_conditioned_data, bins=50, color='blue', alpha=0.7)
plt.title("Ill-Conditioned Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Create a histogram for the logarithmic data
plt.subplot(1, 2, 2)
log_hist, log_bins, _ = plt.hist(log_data, bins=50, color='green', alpha=0.7)
plt.title("Logarithmic Counterpart")
plt.xlabel("Log(Value)")
plt.ylabel("Frequency")

# Find the global maximum (peak) in the log histogram
global_max_index = np.argmax(log_hist)
global_max_value = log_hist[global_max_index]

# Calculate midpoints of the histogram bins for log data
bin_midpoints = 0.5 * (log_bins[:-1] + log_bins[1:])

# Create an interpolated curve for the log histogram using bin midpoints
x_smooth = np.linspace(bin_midpoints.min(), bin_midpoints.max(), 300)
spline = make_interp_spline(bin_midpoints, log_hist, k=3)
y_smooth = spline(x_smooth)

# Plot the interpolated curve for the log histogram and mark the global maximum
plt.subplot(1, 2, 2)
plt.plot(x_smooth, y_smooth, color='red', label='Interpolated Curve')
plt.scatter(bin_midpoints[global_max_index], global_max_value, color='red', marker='o', label='Global Maxima')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 5  # Mean
sigma = 1  # Standard deviation
sample_size = 10000

# Generate random data for the normal distribution
data_normal = np.random.normal(mu, sigma, sample_size)

# Parameters for the normal distribution with logarithmic transformation
log_mu = np.log(mu)  # Logarithm of the mean
log_sigma = 1  # Logarithm of the standard deviation

# Generate random data for the log-normal distribution
data_log_normal = np.random.lognormal(log_mu, log_sigma, sample_size)

# Create a histogram for the normal distribution
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(data_normal, bins=150, density=True, alpha=0.6, color='blue', label='Normal Distribution')
plt.title('Histogram of Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

# Create a histogram for the log-normal distribution
plt.subplot(1, 2, 2)
plt.hist(data_log_normal, bins=150, density=True, alpha=0.6, color='green', label='Log-Normal Distribution')
plt.title('Histogram of Log-Normal Distribution')
plt.xlabel('Value (Logarithmic)')
plt.ylabel('Probability Density')
plt.legend()

# Create PDFs for both distributions
x = np.linspace(0, 2*sigma*mu, 1000)
pdf_normal = norm.pdf(x, mu, sigma)
pdf_log_normal = norm.pdf(np.log(x), log_mu, log_sigma) / x

# Plot PDFs
plt.figure(figsize=(12, 6))
plt.plot(x, pdf_normal, 'b-', label=f'Normal PDF ')
plt.plot(x, pdf_log_normal, 'g-', label=f'Log-Normal PDF')
plt.axvline(x=mu, color='r', linestyle='--', label=f'Normal mean y={mu}')
plt.axvline(x=log_mu, color='r', linestyle='--', label=f'Log-Normal mean x={"{:.2f}".format(log_mu)} => exp => {"{:.2f}".format(np.exp(log_mu))}')
plt.title('Probability Density Functions')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters for the normal distribution
mu = 0  # Mean
sigma = 1  # Standard deviation
sample_size = 10000

# Generate random data for the normal distribution
data_normal = np.random.normal(mu, sigma, sample_size)

# Parameters for the normal distribution with logarithmic transformation
log_mu = np.log(1)  # Logarithm of the mean
log_sigma = 1  # Logarithm of the standard deviation

# Generate random data for the log-normal distribution
data_log_normal = np.random.lognormal(log_mu, log_sigma, sample_size)

# Create PDFs for both distributions
x = np.linspace(-3, 3, 1000)
pdf_normal = norm.pdf(x, mu, sigma)
pdf_log_normal = norm.pdf(np.log(x), log_mu, log_sigma) / x

# Calculate mode and median for both distributions
mode_normal = mu
median_normal = mu
mode_log_normal = np.exp(log_mu - log_sigma**2)
median_log_normal = np.exp(log_mu)

# Plot PDFs
plt.figure(figsize=(12, 6))
plt.plot(x, pdf_normal, 'b-', label='Normal PDF')
plt.plot(x, pdf_log_normal, 'g-', label='Log-Normal PDF (Transformed)')
plt.axvline(x=mode_normal, color='r', linestyle='--', label='Mode (Normal)')
plt.axvline(x=median_normal, color='m', linestyle='--', label='Median (Normal)')
plt.axvline(x=mode_log_normal, color='y', linestyle='--', label='Mode (Log-Normal)')
plt.axvline(x=median_log_normal, color='c', linestyle='--', label='Median (Log-Normal)')

# Add values to the lines
plt.text(mode_normal + 0.05, 0.4, f'{mode_normal:.2f}', color='r', fontsize=12)
plt.text(median_normal - 0.3, 0.2, f'{median_normal:.2f}', color='m', fontsize=12)
plt.text(mode_log_normal + 0.1, 0.4, f'{mode_log_normal:.2f}', color='y', fontsize=12)
plt.text(median_log_normal - 0.3, 0.2, f'{median_log_normal:.2f}', color='c', fontsize=12)

plt.title('Probability Density Functions with Mode and Median')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()

plt.show()
```
