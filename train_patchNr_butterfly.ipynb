```python
import sqlite3

import matplotlib.pyplot as plt
```

```python
#TABLE flow_model_train_loss(step, loss)
version = 1
loss_db_path = f'results/patch_nr/butterfly/version_{version}/loss.db'
```

```python
connection = sqlite3.connect(loss_db_path)
cursor = connection.cursor()
```

```python
cursor.execute('SELECT * from flow_model_train_loss')
step, loss =zip(*cursor.fetchall())
cursor.execute('SELECT * from flow_model_validation_loss')
val_step, val_loss =zip(*cursor.fetchall())
```

```python
plt.plot(step, loss, label='Training Loss', color='lightcoral')
plt.plot(val_step, val_loss, label='Validation Loss', color='red')
plt.ylabel('Loss')
plt.xlabel('Iterations')
plt.legend()
plt.show()
```
