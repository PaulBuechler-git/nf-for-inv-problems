```python
from matplotlib import pyplot as plt
import numpy as np
from dataset.FastPatchExtractor import FastPatchDataset
import torch
import torchvision as tv
```

```python
train_img_path = '../data/set12/train'
p_dims = 6
train_data = FastPatchDataset(train_img_path, p_dims)
```

```python
images_len=len(train_data.images)
fig = plt.figure(figsize=(len(train_data.images), 1))
for i in range(1, images_len):
    fig.add_subplot(1, images_len, i)
    plt.imshow(train_data.images[i].squeeze(),cmap='gray')
plt.show()
```

```python
batch_size = 32
batch = torch.reshape(train_data.get_batch(batch_size), (batch_size, p_dims, p_dims))
cols = batch_size//2
rows = 2
fig = plt.figure(figsize=(batch_size//2, 2.5))
for i in range(1, cols*rows +1):
    fig.add_subplot(rows, cols, i)
    plt.imshow(torch.randint(0, 9, (1, p_dims, p_dims)).squeeze(),cmap='gray')
plt.show()
```
