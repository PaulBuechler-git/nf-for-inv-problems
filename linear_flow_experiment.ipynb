```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
covariance = [[1, 0], [0, 1]]
num_samples = 10000
samples = np.random.multivariate_normal(mean, covariance, num_samples)
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(samples[:, 0], samples[:, 1], s=5, alpha=0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
A = np.array([[1.5, 0], [1, -1]])
b = [5, 1]
transformed_samples = np.array([A @ s +b for s in samples])
```

```python
plt.figure(figsize=(4, 4))
plt.scatter(transformed_samples[:, 0], transformed_samples[:, 1], s=5, alpha=0.6)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
path = 'images/linear_flow.png'
```

```python
fig, axes = plt.subplots(1, 2, figsize=(5,3))
axes[0].scatter(samples[:, 0], samples[:, 1], s=5, alpha=0.6)
axes[1].scatter(transformed_samples[:, 0], transformed_samples[:, 1], s=5, alpha=0.6)
fig.tight_layout()
fig.show()
fig.savefig(path)
```
